```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head(30)
```

```python
df.info()
```

```python
import math
Age_new = math.floor(df['Age'].mean())
```

```python
Age_new
```

```python
df.Age = df.Age.fillna(Age_new)


x=df[['Pclass','Sex','Age','Fare']]
y=df.Survived
```

```python
x.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()

```

```python
x['Sex_n'] = le_Sex.fit_transform(x['Sex'])
```

```python
x.head(10)
```

```python
x_new=x[["Pclass","Age","Fare","Sex_n"]]
```

```python
x_new.head()
```

```python
y.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
x_new_train,x_new_test,y_train,y_test=train_test_split(x_new,y, test_size=0.2)
```

```python
model = DecisionTreeClassifier(criterion="entropy")
model.fit(x_new_train,y_train)
```

```python
predictions=model.predict(x_new_test)
predictions
score=accuracy_score(y_test,predictions)
score
```
